```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados
df = pd.read_csv('SINASC_RO_2019.csv')

# Estilo dos gráficos
sns.set(style='whitegrid')
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['IDADEMAE'], color='skyblue')
plt.title('Boxplot da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.tight_layout()
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['PESO'], color='lightgreen')
plt.title('Boxplot do Peso do Bebê')
plt.xlabel('Peso (em gramas)')
plt.tight_layout()
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['IDADEMAE'], bins=20, color='steelblue', kde=True)
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.tight_layout()
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['PESO'], bins=30, color='seagreen', kde=True)
plt.title('Histograma do Peso do Bebê')
plt.xlabel('Peso (em gramas)')
plt.ylabel('Frequência')
plt.tight_layout()
plt.show()
```
