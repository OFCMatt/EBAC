1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import ttest_ind

df = pd.read_csv("previsao_de_renda.csv")
````

````python
# Média de renda por posse de imóvel
print(df.groupby('posse_de_imovel')['renda'].mean())

# Média de renda por posse de veículo
print(df.groupby('posse_de_veiculo')['renda'].mean())
````

````python
# Boxplot para posse de imóvel
sns.boxplot(x='posse_de_imovel', y='renda', data=df)
plt.title('Renda por Posse de Imóvel')
plt.show()

# Boxplot para posse de veículo
sns.boxplot(x='posse_de_veiculo', y='renda', data=df)
plt.title('Renda por Posse de Veículo')
plt.show()
````

````python
# Comparação entre quem tem e quem não tem imóvel
renda_imovel_sim = df[df['posse_de_imovel'] == True]['renda']
renda_imovel_nao = df[df['posse_de_imovel'] == False]['renda']

t_stat_imo, p_val_imo = ttest_ind(renda_imovel_sim, renda_imovel_nao, equal_var=False)
print(f"p-valor (posse de imóvel): {p_val_imo:.4f}")

# Comparação entre quem tem e quem não tem veículo
renda_veiculo_sim = df[df['posse_de_veiculo'] == True]['renda']
renda_veiculo_nao = df[df['posse_de_veiculo'] == False]['renda']

t_stat_veic, p_val_veic = ttest_ind(renda_veiculo_sim, renda_veiculo_nao, equal_var=False)
print(f"p-valor (posse de veículo): {p_val_veic:.4f}")
````
